```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Expanded_data_with_more_features.csv")
df.info()
print(df.head())
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# remove column which isn't named
df=df.drop("Unnamed: 0" , axis=1)
print(df.head())
```

```python
#gender distribution
plt.figure(figsize=(6,5))
fig=sns.countplot(data = df, x="Gender")
fig.bar_label(fig.containers[0],fontsize=9,color="green")
plt.title("Gender Distribution",fontsize=24,color="red")
plt.show()

#no of females is more than no. of males
```

```python
group= df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(group)
group1= df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(group1)
```

```python
#impact of parents martial status on student's score
plt.figure(figsize=(4,4))
sns.heatmap(group, annot=True)
plt.title("impact of parents martial status on student's score",fontsize=12,color="blue")
plt.show()
#their is negligible impact 
```

```python
#impact of parent's education on student's score
plt.figure(figsize=(4,5))
sns.heatmap(group1, annot=True)
plt.title("Impact of parent's education on student's score",fontsize=12,color="blue")
plt.show()
#eduction of parents have a good impact on their students education
```
